Задание 1: Регулярные выражения

1.    Напишите регулярное выражение, которое найдет все слова, которые начинаются с сочетания букв "**ста**".

2.    Напишите регулярное выражение, которое найдет все слова с сочетанием букв **ста** как в начале, так и внутри слов.

3. Модернизируйте это выражение так, чтобы буква **т** была опциональной (вместо нее может быть любой другой символ).



*Не забывайте импортировать нужные библиотеки*

```python
# 1) регулярное выражение, которое найдет все слова, которые начинаются с сочетания букв "ста"
import re
print(re.findall(r'\b[Сс]та\w*', text))
```

```python
# 2) регулярное выражение, которое найдет все слова с сочетанием букв ста как в начале, так и внутри слов
print(re.findall(r'\w*[Сс]та\w*', text))
```

```python
# 3) чтобы буква т была опциональной (вместо нее может быть любой другой символ)
print(re.findall(r'\w*[Сс][а-я]а\w*', text))
```

Задание 2: Нормализация текста

1.   Преобразуйте текст из переменной text к нижнему регистру.

2.   Удалите все символы пунктуации и переноса строки (\n), используя регулярные выражения.


```python
# 1) приводим к нижнему регистру
lower_text = text.lower()
print(lower_text)
```

```python
# 2) удаляем символы пунктуации и переноса строки
lower_no_n = re.sub(r'\n', ' ', lower_text) # заменяем перенос строки на пробел
lower_no_punkt = re.sub(r'[^\w\s]', '', lower_no_n) # удаляем пунктуацию
print(lower_no_punkt)
```

```python
# 1.1 токенизация нормализованного текста через библиотеку nltk
import nltk # импортируем библиотеку
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
from nltk import word_tokenize # импортируем из библиотеки функцию для токенизации
```

```python
tokens = word_tokenize(lower_no_punkt) # токенизация нормализованного текста через библиотеку nltk
print(tokens)
print('Количество токенов', len(tokens))
print('Количество уникальных токенов', len(set(tokens)))
```

```python
# 2 токенизация изначального текста на строки при помощи split('\n'). Отобразите сколько их
tok_lines = text.split('\n')
print(tok_lines)
print('Количество токенов по переносу строки', len(tok_lines))
```

```python
# 3) соберём все токены с заглавной буквы
toks = word_tokenize(text) # токенизируем текст библиотекой
print(toks)
print('Количество токенов в тексте', len(toks))
cap_toks = [] # создаем пустой список для токенов с заглавной буквы
for tok in range(len(toks)): # цикл для токенов
  if toks[tok][0].isupper(): # проверяем, является ли заглавным [0] символ каждого токена
    cap_toks.append(toks[tok]) # добавляем в список токенов с заглавной буквы
print(cap_toks)
print('Количество токенов с заглавной буквы', len(cap_toks))
```

Задание 4: Удаление стоп слов


1. Импортируйте список стоп-слов для русского языка из NLTK.
2. Напишите цикл, который пройдет по всем токенам и вернет новый список токенов без стоп-слов.
3. Сколько слов осталось в тексте? А уникальных?
4. *При помощи регулярного выражения найдите все слова с сочетанием букв 'бол' (с опциональной о) внутри слова. Добавьте их к стоп-словам и удалите из текста. Сколько осталось слов в тексте?




```python
from nltk.corpus import stopwords # импортируем модуль stopwords из nltk
```

```python
# 1) импортируем список стоп-слов для русского языка в nltk
nltk_stopwords_ru = stopwords.words('russian')
print(nltk_stopwords_ru) # посмотрим на него
```

```python
# 2) циклом for соберём токены, которые не являются стоп-словами
tokens_no_stop_words = []
for token in word_tokenize(text.lower()):
    if token not in nltk_stopwords_ru:
        tokens_no_stop_words.append(token)
print(tokens_no_stop_words)
```

```python
print('Количество токенов без стоп-слов', len(tokens_no_stop_words))
print('Количество уникальных токенов без стоп-слов', len(set(tokens_no_stop_words)))
```
